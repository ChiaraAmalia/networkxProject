```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
import pandas as pd
import matplotlib.colors as colors
```

```python
pd.set_option('display.max_rows', 10)
Relations = pd.read_csv('R_M_relations.csv')
```

```python
Graphtype = nx.Graph()
graph = nx.from_pandas_edgelist(Relations,source = 'src', target = 'dest', create_using=Graphtype)
```

Group degree centrality

```python
gcentral_bet = []
group1 ={'Rick Sanchez','Diane Sanchez','Leonard Smith','Joyce Smith','Space Beth','Beth Smith','Jerry Smith','Summer Smith','Morty Smith','Sticky','Naruto Smith'}
gcentral_bet.append(nx.group_degree_centrality(graph,group1))
print(gcentral_bet[0])

group2 ={'Rick Sanchez','Diane Sanchez','Leonard Smith','Joyce Smith','Space Beth','Beth Smith','Jerry Smith','Summer Smith','Morty Smith','Gwendolyn','Morty Jr.','Sperm Queen'}
gcentral_bet.append(nx.group_degree_centrality(graph,group2))
print(gcentral_bet[1])

group3 ={'Rick Sanchez','Diane Sanchez','Poñeta','Beth Sanchez (C-137)','Rick\'s Foal','Chud King'}
gcentral_bet.append(nx.group_degree_centrality(graph,group3))
print(gcentral_bet[2])
plt.figure(figsize=(15,10))
sns.barplot(x=['destro','centro','sinisto'], y=gcentral_bet)
```
