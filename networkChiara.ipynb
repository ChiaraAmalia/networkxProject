```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
import pandas as pd
import matplotlib.colors as colors
```

```python
pd.set_option('display.max_rows', 10)
Relations = pd.read_csv('R_M_relations.csv')
```

```python
Graphtype = nx.Graph()
graph = nx.from_pandas_edgelist(Relations,source = 'src', target = 'dest', create_using=Graphtype)
```

GRAFO INIZIALE

```python
pos = nx.spring_layout(graph)
plt.figure(figsize=(20,10))
nx.draw(graph, pos, with_labels=True)
plt.show()
```

```python
print(nx.info(graph))
```

```python
S = [graph.subgraph(c).copy() for c in nx.connected_components(graph)]
graph = S[0]
```

```python
pos = nx.spring_layout(graph)
plt.figure(figsize=(20,10))
nx.draw(graph, pos, with_labels=True)
plt.show()
```

```python
print("NODI: ", len(graph.nodes))
print("ARCHI: ", len(graph.edges))
print("DENSITA", nx.density(graph))
print("RAGGIO: ", nx.radius(graph))
print("DIAMETRO: ", nx.diameter(graph))
print("PERIFERIA: ", nx.periphery(graph))
print("CLUSTERING: %0.3f" % nx.average_clustering(graph))
print("IS CONNECTED: ", nx.is_connected(graph))
```
